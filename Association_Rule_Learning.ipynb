```python
market = []
for i in range (0 , df.shape[0]):
    customer = []
    for j in df.columns:
        if type(df[j][i])==str:
            customer.append(df[j][i])
    market.append(customer)
    
#  this is done to remove the NAN values
 
```

MODEL MAKING STARTS HERE

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
tr.transform(market)
```

```python
df= pd.DataFrame(tr.transform(market) , columns=tr.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori 
```

```python
apriori(df , min_support=0.07 , use_colnames=True , max_len=3).sort_values(by=["support"])
```
