```python
# Ordinal Encoding

import pandas as pd
```

```python
df =pd.DataFrame({"Size":["s", "m" , "l" , "xl" , "s" , "m" , "l" , "s","s","s","l","xl","m" ]})
df.head(4)
```

```python
ord_data = [["s" , "m", "l" , "xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
df["Size_en"] = oe.fit_transform(df[["Size"]])
```

```python
df
```

```python
new_ord_data = {"s": 0, "m": 3, "l": 2, "xl": 1, "xxl": 4, }
```

```python
df["Size_en_map"] = df["Size"].map(new_ord_data)
```
