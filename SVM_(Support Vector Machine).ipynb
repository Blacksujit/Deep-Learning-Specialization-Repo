```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , LogisticRegression
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("polynomial_classification_dataset.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(4,3))

sns.scatterplot(x="X1" , y="X2" , data=df , hue="Class")

plt.show()
```

```python
x = df.iloc[:,:-1]
y = df["Class"]
```

```python
# Model Making Starts here
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, random_state=15)
```

```python
from sklearn.svm import SVC 
# beacuse data is of classification
```

```python
sv = SVC(kernel='rbf')
sv.fit(x_train , y_train)

# there are different kernels try that to seperate the data and for the same observe the graph
```

```python
sv.score(x_test , y_test)*100
```

```python
sv.score(x_train , y_train)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=sv)
plt.show()
```
